### Stemming

It is only sensitive to the suffix.
Porter Stemmer
Lancaster stemmer. Find the difference. ONe is string for one word while the other is not. They have the list of suffix words

```python
from nltk.stem import PorterStemmer
stemmerporter = PorterStemmer()
stemmerporter.stem('happiness')
```

```python
from nltk.stem import LancasterStemmer
stemmerporter = LancasterStemmer()
stemmerporter.stem('happiness')
```

```python
from nltk.stem import RegexpStemmer
stemmerregexp = RegexpStemmer('happen')
stemmerregexp.stem('happening')
```

```python
# Find how many languages does snowball support
from nltk.stem import SnowballStemmer
SnowballStemmer.languages
frenchstemmer = SnowballStemmer('spanish')
frenchstemmer.stem('correr')
```

```python
#texts = "World War I[j] or the First World War (28 July 1914 – 11 November 1918) was a global conflict fought between two coalitions: the Allies and the Central Powers. Fighting took place throughout Europe, the Middle East, Africa, the Pacific, and parts of Asia. One of the deadliest wars in history, it resulted in an estimated 9 million soldiers dead and 23 million wounded, plus another 5 million civilian deaths from various causes. Millions more died as a result of genocide, and the war was a major factor in the 1918 Spanish flu pandemic."
texts = "In early 1917, the United States entered the war on the side of the Allies, and later the same year, the Bolsheviks seized power in the Russian October Revolution, making peace with the Central Powers in early 1918. Germany launched an offensive in the west in March 1918, and despite initial success, it left the German Army exhausted and demoralised. A successful Allied counter-offensive later that year caused a collapse of the German frontline. By the end of 1918, Bulgaria, the Ottoman Empire and Austria-Hungary agreed to armistices with the Allies, leaving Germany isolated. Facing revolution at home and with his army on the verge of mutiny, Kaiser Wilhelm II abdicated on 9 November."
sentences = nltk.sent_tokenize(texts)
print(sentences)
for text in texts: 
    #sentences = nltk.sent_tokenize(text)
    #print(sentences)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        print(words)
        tagged = nltk.pos_tag(words)
        print(tagged)
```
